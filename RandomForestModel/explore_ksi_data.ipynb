```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
import plotly.express as px
import numpy as np
```

```python
df = pd.read_csv("ksi_collisions.csv")
severe_df = df[df["SEVERE_COLLISION"] == 1]
```

```python
sns.countplot(x="YEAR", data=df, hue="SEVERE_COLLISION")
plt.xticks(rotation=45) 
plt.tight_layout()     
plt.show()
```

```python
sns.countplot(x="HOUR", data=df)
```

```python
sns.countplot(x="HOUR", data=severe_df)
```

```python
plt.figure(figsize=(8,5))
sns.countplot(x='TIME_OF_DAY', hue='SEVERE_COLLISION', data=df, palette='coolwarm')
plt.title('Collision Severity by Time Of Day')
plt.xlabel('Time Of Day')
plt.ylabel('Count')
plt.legend(title='Severe Collision', labels=['No (0)', 'Yes (1)'])
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.countplot(x='LIGHT', hue='SEVERE_COLLISION', data=df, palette='coolwarm')
plt.title('Collision Severity by LIGHT')
plt.xlabel('Light Level')
plt.ylabel('Count')
plt.legend(title='Severe Collision', labels=['No (0)', 'Yes (1)'])
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.countplot(x='VISIBILITY', hue='SEVERE_COLLISION', data=df, palette='coolwarm')
plt.title('Collision Severity by VISIBILITY')
plt.xlabel('Visibility')
plt.ylabel('Count')
plt.legend(title='Severe Collision', labels=['No (0)', 'Yes (1)'])
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.countplot(x='SEASON', hue='SEVERE_COLLISION', data=df, palette='coolwarm')
plt.title('Collision Severity by Season')
plt.xlabel('Season')
plt.ylabel('Count')
plt.legend(title='Severe Collision', labels=['No (0)', 'Yes (1)'])
plt.show()
```
